```python
import geopandas as gpd
from libpysal.weights import Queen

# Step 1: Read the GeoJSON file
geojson_file = '/content/drive/MyDrive/London_heat/selected_MSOASVI2.geojson'
gdf = gpd.read_file(geojson_file)

# Step 2: Create Queen's spatial weights
weights = Queen.from_dataframe(gdf)
```

```python
source = []
target = []

for key, neighbors in weights.neighbors.items():
    for neighbor in neighbors:
        source.append(key)
        target.append(neighbor)

# Step 4: Create a DataFrame and save to CSV
connections = pd.DataFrame({'source': source, 'target': target})
connections.to_csv('/content/drive/MyDrive/London_heat/spatial_weights_connections.csv', index=False)

print("CSV file 'spatial_weights_connections.csv' created successfully.")
```

```python
import geopandas as gpd
import pandas as pd

# Step 1: Read the polygon and line GeoJSON files
polygon_file = '/content/drive/MyDrive/London_heat/selected_MSOASVI2.geojson'
# line_file = '/content/drive/MyDrive/London_heat/waterway.geojson'
line_file = '/content/drive/MyDrive/London_heat/ABRoads.geojson'


polygons = gpd.read_file(polygon_file)
lines = gpd.read_file(line_file)
if polygons.crs != lines.crs:
    lines = lines.to_crs(polygons.crs)
    print("Adjusted Line CRS to match Polygon CRS")
# Step 2: Perform a spatial join to find intersections
lines_in_polygons = gpd.sjoin(lines, polygons, how="left", op="intersects")

lines_in_polygons
```

```python
lines_in_polygons['MSOA_CODE']
```

```python
# Step 3: Determine which polygons are connected by the same line
# Create a DataFrame with line IDs and the polygons they intersect
connections = lines_in_polygons[['fid', 'MSOA_CODE']].dropna()
connections.columns = ['fid', 'MSOA_CODE']

# Create a dictionary to map lines to the polygons they connect
line_to_polygons = connections.groupby('fid')['MSOA_CODE'].apply(list).to_dict()

# Create a set to store the unique connections
polygon_connections = set()

# Iterate through the dictionary to create pairs of connected polygons
line_to_polygons
# connections
```

```python
for polygons in line_to_polygons.values():
    if len(polygons) > 1:
        for i in range(len(polygons)):
            for j in range(i + 1, len(polygons)):
                polygon_connections.add((polygons[i], polygons[j]))

# Convert the set of connections to a DataFrame
connections_df = pd.DataFrame(polygon_connections, columns=['source', 'target'])

# Step 4: Save the connections to a CSV file
connections_df.to_csv('/content/drive/MyDrive/London_heat/ABRoads_connections.csv', index=False)

print("CSV file 'polygon_connections.csv' created successfully.")
```

```python
import pandas as pd
csv_file = '/content/drive/MyDrive/London_heat/LOAC_MSOA.csv'
df = pd.read_csv(csv_file)

# Define the columns for common values and areas
class_column = 'Super'  # replace with the name of the column with common values
area_column = 'MSOA_CODE'  # replace with the name of the column with area names or IDs

# Step 2: Group by the area column and count the occurrences of each class
class_counts = df.groupby([area_column, class_column]).size().reset_index(name='counts')

predominant_class = class_counts.loc[class_counts.groupby(area_column)['counts'].idxmax()]

# Only keep the area and the predominant class
predominant_class = predominant_class[[area_column, class_column]]

# Rename columns for clarity
predominant_class.columns = [area_column, 'predominant_class']

# Step 4: Save the results to a new CSV file
predominant_class.to_csv('predominant_class_by_area.csv', index=False)

print("CSV file 'predominant_class_by_area.csv' created successfully.")
```

```python
import pandas as pd

# Step 1: Read the CSV file
csv_file = 'predominant_class_by_area.csv'
df = pd.read_csv(csv_file)

# Define the columns for common values and areas
common_column = 'predominant_class'  # replace with the name of the column with common values
area_column = 'MSOA_CODE'  # replace with the name of the column with area names or IDs

# Step 2: Group by the common values and create connections
grouped = df.groupby(common_column)[area_column].apply(list).reset_index()

# Step 3: Create connections for the areas
connections = []

for areas in grouped[area_column]:
    if len(areas) > 1:
        for i in range(len(areas)):
            for j in range(i + 1, len(areas)):
                connections.append((areas[i], areas[j]))

# Convert the connections list to a DataFrame
connections_df = pd.DataFrame(connections, columns=['source', 'target'])

# Step 4: Save the connections to a CSV file
connections_df.to_csv('/content/drive/MyDrive/London_heat/LOAC_area_connections.csv', index=False)

print("CSV file 'LOAC_area_connections.csv' created successfully.")
```

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

# Load points data from CSV
points_df = pd.read_csv('/content/drive/MyDrive/London_heat/london_SVI_points.csv')

# Assuming your points data has 'longitude' and 'latitude' columns
geometry = [Point(xy) for xy in zip(points_df['x'], points_df['y'])]
points_gdf = gpd.GeoDataFrame(points_df, geometry=geometry)
points_gdf.set_crs(epsg=4326, inplace=True)

# Load polygons data (shapefile or another geospatial format)
polygons_gdf = gpd.read_file('/content/drive/MyDrive/London_heat/selected_MSOASVI2.geojson')

# Ensure the coordinate reference systems (CRS) match
if points_gdf.crs != polygons_gdf.crs:
    points_gdf = points_gdf.to_crs(polygons_gdf.crs)
    print("Transformed CRS of points GeoDataFrame to match polygons GeoDataFrame:", points_gdf.crs)
# Perform spatial join to map points to polygons
joined_gdf = gpd.sjoin(points_gdf, polygons_gdf, how="left", op="within")

# Calculate mean values of the points' features based on each polygon
# Assuming 'feature1' and 'feature2' are the columns for which you want to calculate the mean
```

```python
result_df = joined_gdf.drop(columns='geometry')
result_df = result_df[['MSOA_CODE','road','sidewalk','building','wall',	'fence',	'pole',	'traffic light',	'traffic sign',	'vegetation',	'terrain',	'sky',	'person',	'rider',	'car',	'truck',	'bus',	'train',	'motorcycle',	'bicycle']]
```

```python
mean_values = result_df.groupby('MSOA_CODE').mean()[['road','sidewalk','building','wall',	'fence',	'pole',	'traffic light',	'traffic sign',	'vegetation',	'terrain',	'sky',	'person',	'rider',	'car',	'truck',	'bus',	'train',	'motorcycle',	'bicycle']]
mean_values
```

```python
# Merge mean values back to polygons GeoDataFrame
polygons_gdf = polygons_gdf.join(mean_values, on='MSOA_CODE')

# Save the result to a new shapefile or another format if needed
result_df = polygons_gdf.drop(columns='geometry')  # Drop the geometry column if you don't need it in the CSV
result_df.to_csv('/content/drive/MyDrive/London_heat/Connections/data_features.csv', index=False)

print("Mean values calculated and saved to CSV file:")
print(result_df)
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point

# Load points data from CSV
points_df = pd.read_csv('/content/drive/MyDrive/London_heat/london_SVI_points.csv')

# Assuming your points data has 'longitude' and 'latitude' columns
geometry = [Point(xy) for xy in zip(points_df['x'], points_df['y'])]
points_gdf = gpd.GeoDataFrame(points_df, geometry=geometry)
points_gdf.set_crs(epsg=4326, inplace=True)
# Load polygons data from GeoJSON
polygons_gdf = gpd.read_file('/content/drive/MyDrive/London_heat/selected_MSOASVI2.geojson')

# Check and print the CRS of both GeoDataFrames
print("CRS of points GeoDataFrame:", points_gdf.crs)
print("CRS of polygons GeoDataFrame:", polygons_gdf.crs)

# Transform the CRS of the points to match the CRS of the polygons if they are different
if points_gdf.crs != polygons_gdf.crs:
    points_gdf = points_gdf.to_crs(polygons_gdf.crs)
    print("Transformed CRS of points GeoDataFrame to match polygons GeoDataFrame:", points_gdf.crs)

# Plotting the polygons and points
fig, ax = plt.subplots(figsize=(10, 10))
polygons_gdf.plot(ax=ax, color='lightgrey', edgecolor='black')
points_gdf.plot(ax=ax, color='red', markersize=5)

# Customize the plot
plt.title('Points Mapped onto Polygons')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Save the plot as an image file
plt.savefig('mapped_points.png')

# Show the plot
plt.show()
```
